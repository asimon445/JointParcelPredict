```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import itertools

# Load the .mat file
mat_data = scipy.io.loadmat('/Users/ajsimon/Documents/Data/Constable_lab/Functional_parcellations/figures/subvolumes_L.mat')

# Assuming the variable in the .mat file is named 'subvolumes'
subvolumes = mat_data['new_subvolumes']
```

```python
subvolumes[2]
```

```python
# Set up the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a color cycle iterator for different colors
colors = itertools.cycle(plt.cm.tab20.colors)

# Plot each subvolume in a different color
for i in range(subvolumes.shape[1]):
    subvolume_data = subvolumes[0,i]  # Extract data from cell
    x = subvolume_data[:, 0]
    y = subvolume_data[:, 1]
    z = subvolume_data[:, 2]
    
    color = next(colors)
    ax.scatter(x, y, z, color=color, label=f'Subvolume {i+1}', s=1)  # s=1 for small marker size

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot of Subvolumes')

# Rotate the plot programmatically (optional)
ax.view_init(elev=10, azim=-5)  # Adjust elev (up/down) and azim (left/right) angles

# Move the legend outside the plot
ax.legend(loc='upper left', bbox_to_anchor=(1.5, 1), borderaxespad=0.)
ax.grid(False)
ax.w_xaxis.pane.set_facecolor('white')
ax.w_yaxis.pane.set_facecolor('white')
ax.w_zaxis.pane.set_facecolor('white')

# Show plot
plt.show()
```
